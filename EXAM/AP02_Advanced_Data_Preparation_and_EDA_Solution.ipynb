#### <b>Aufgabe (2): Extrahieren Sie die numerischen Werte der folgenden Variablen, und speichern Sie diese in separaten Variablen im DataFrame ab.</b>
<b>Details zur Aufgabenstellung:</b>
- Extrahieren Sie die numerischen Werte der folgenden Variablen, und speichern Sie diese in separaten Variablen ab:
  - price_raw
  - gear_count_raw
  - battery_life_raw
  - frame_size_raw
  - wheel_size_raw<br>
- Benennen Sie die Spalten der neuen Variablen als (in Klammern ist der Datentyp angegeben):
  - price (Int64)
  - gear_count (Int64)
  - battery_life (Int64)
  - frame_size (Int64)
  - wheel_size (Float64)
- Stellen Sie ein Subset des DataFrames nur mit den neuen Spalten und den ersten 5 Zeilen im Jupyter Notebook dar.<br><br>

- Tipp: Die folgenden Code Zeilen inkl. regular expressions sind eine Hilfestellung zum Lösen der Aufgabe:
  ```python
    d1 = re.findall("CHF\s*(\d+)\.-", i)
    d1 = re.findall("(\d+)\s*Gänge", i)
    d1 = re.findall("(\d+)\s*km", str(i))
    d1 = re.findall("(\d+)\s*cm", i)
    d1 = re.findall("(\d+(?:[.,]\d+)?)", i)
  ```
- Gross- und Kleinschreibung beim Abspeichern als neue Variable beachten, z.B.:
  - df['my_int_var'] = pd.Series(my_int_var, dtype="Int64") -> kann mit NAs umgehen
  - df['my_float_var'] = pd.Series(my_float_var, dtype="Float64") -> kann mit NAs umgehen
  
<b style="color: gray;">(max. erreichbare Punkte: 8)</b>

````python
# Extrahieren der numerischen Werte aus 'price_raw' und speichern in separaten Variablen
price = []
for i in df['price_raw']:
    d1 = re.findall("CHF\s*(\d+)\.-", i)
    try:
        d2 = int(''.join(d1))
    except:
        d2 = None
    price.append(d2)

# Save as new variable in the pandas data frame
df['price'] = pd.Series(price, dtype="Int64")
````

````python
# Extrahieren der numerischen Werte aus 'gear_count_raw' und speichern in separaten Variablen
gear_count = []
for i in df['gear_count_raw']:
    d1 = re.findall("(\d+)\s*Gänge", i)
    try:
        d2 = int(''.join(d1))
    except:
        d2 = None
    gear_count.append(d2)

# Save as new variable in the pandas data frame
df['gear_count'] = pd.Series(gear_count, dtype="Int64")
````

````python
# Extrahieren der numerischen Werte aus 'battery_life_raw' und speichern in separaten Variablen
battery = []
for i in df['battery_life_raw']:
    d1 = d1 = re.findall("(\d+)\s*km", str(i))
    try:
        d2 = int(''.join(d1))
    except:
        d2 = None
    battery.append(d2)

# Save as new variable in the pandas data frame
df['battery_life'] = pd.Series(battery, dtype="Int64")
````

````python
# Extrahieren der numerischen Werte aus 'frame_size_raw' und speichern in separaten Variablen
frame_size = []
for i in df['frame_size_raw']:
    d1 = re.findall("(\d+)\s*cm", i)
    try:
        d2 = int(''.join(d1))
    except:
        d2 = None
    frame_size.append(d2)

# Save as new variable in the pandas data frame
df['frame_size'] = pd.Series(frame_size, dtype="Int64")
````

````python
# Extrahieren der numerischen Werte aus 'wheel_size_raw' und speichern in separaten Variablen (Tipp: floating point anstatt integer Variable)
wheel_size = []
for i in df['wheel_size_raw']:
    d1 = re.findall("(\d+(?:[.,]\d+)?)", i)
    try:
        d2 = float(''.join(d1))
    except:
        d2 = None
    wheel_size.append(d2)

# Save as new variable in the pandas data frame
df['wheel_size'] = pd.Series(wheel_size, dtype="Float64")
````

````python
# Einlesen der aufbereiteten Daten (damit alle auf dem gleichen Stand sind)
engine = create_engine('sqlite:///./Data/bicycle_data_prepared.db')
df = pd.read_sql_query("""SELECT * FROM bicycle_data_prepared""", con=engine)
engine.dispose()

# Columns der relevanten numerischen Variablen auswählen
cols = ['price', 'gear_count', 'battery_life', 'frame_size', 'wheel_size']

# Berechnung der statistischen Kennzahlen der numerischen Spalten in 'df'
print('Summary Statistic:\n', df[cols].describe().round(2))

# Separate Berechnung der Skewness der numerischen Spalten
print("\nSkewness:")
for col in cols:
    print(f"Skewness of {col}:", df[col].skew().round(4))
````

#### <b>Aufgabe (4): Erstellen Sie eine Pivot-Tabelle mit den Mittwelwerten der numerischen Variablen pro Hersteller.</b>
<b>Details zur Aufgabenstellung:</b>
- Die Angabe zum Hersteller finden Sie in der Variable 'make'.
- Verwenden Sie die für die Mittelwertsberechnung die Variablen 'price', 'gear_count','frame_size' und 'wheel_size'.
- Stellen Sie die Pivot-Tabelle im Jupyter Notebook dar.

<b style="color: gray;">(max. erreichbare Punkte: 4)</b>

#### <b>Aufgabe (7): Erstellen Sie ein Barchart mit der Anzahl Bikes pro Hersteller für alle Bikes, welche <u>keine</u> E-Bikes sind.</b>
<b>Details zur Aufgabenstellung:</b>
- Erstellen Sie ein Subset der Daten nur mit den Bikes, welche keine E-Bikes sind.
- Die Information, ob es sich um ein E-Bike handelt finden Sie in der Spalte 'is_ebike' (0 = 'kein E-Bike', 1 = 'E-Bike').
- Ergänzen Sie den Python Code unten dort wo notwendig.
- Die Angabe zum Hersteller finden Sie in der Variable 'make'.
- Im Barchart müssen die Hersteller auf der vertikalen und die Anzahl Bikes auf der horizontalen Achse abgebildet sein.
- Im Barchart müssen die Anzahl Bikes pro Hersteller der Grösse nach absteigend sortiert sein.
- Beschriften Sie die Achsen des Barcharts und geben Sie dem Barchart einen Titel.

Tipp: Ein Subset mit den Bikes, welche keine E-Bikes sind, kann mit Hilfe von df[df['is_ebike'] == 0] erstellt werden.

<b style="color: gray;">(max. erreichbare Punkte: 8)</b>

````python
# Subset der Daten nur mit den Bikes, welche keine E-Bikes sind
df_no_ebike = df[df['is_ebike'] == 0]

# Gruppierung der Daten nach 'make' und Berechnung der Anzahl Bikes pro Hersteller
df_bar = df_no_ebike['make'].value_counts().sort_values(ascending=True)

# Zusammenstellen der Werte für den Barchart
napart = list(df_bar.values)
index = list(df_bar.index.values)
y_pos = np.arange(len(index))

# Erstellen des Barcharts
fig, ax = plt.subplots(figsize=(7,4))
ax.barh(y_pos, napart, align='center', color='steelblue')
ax.set_yticks(y_pos, index)
ax.set_xlabel('Anzahl Bikes', fontsize=10)
ax.set_ylabel('Make', fontsize=10)
ax.set_xlim(0, 600)
ax.set_title('Anzahl Bikes pro Hersteller', fontsize=12)

# Anzeigen der Graphik
plt.show()
````
